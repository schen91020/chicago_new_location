```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline 
```

```python
pf = pd.read_csv('CTA_-_Ridership_-__L__Station_Entries_-_Daily_Totals.csv', header = 0, parse_dates='date')
```

```python
pf.head()
```

```python
station_names=set(pf['stationname'])
len(station_names)
```

```python
rides_by_station=pf.groupby('stationname').sum()
```

```python
rides_by_station.loc[rides_by_station['rides'].idxmax()]
```

```python
rides_by_station.loc[rides_by_station['rides'].idxmin()]
```

```python
pf_std=pf[pf['daytype'] == 'W'].groupby('station_id').std()
pf_std[pf_std['rides'] == max(pf_std['rides'])]
```

```python
max_std_pf = pf[pf['station_id'].isin([41660])]
max_std_pf.plot(x = 'date', y = 'rides', figsize=(20, 5))
```

```python
pf_std=pf[pf['daytype'] == 'W'].groupby('station_id').std()
pf_std[pf_std['rides'] == min(pf_std['rides'])]
```

```python
min_std_pf = pf[pf['station_id'].isin([41580])]
min_std_pf.plot(x = 'date', y = 'rides', figsize=(20, 5))
```

```python
pf_ct=pf[pf['daytype'] == 'W'].groupby('station_id').sum()
pf_ct[pf_std['rides'] == max(pf_std['rides'])]
```

```python
max_sum_pf = pf[pf['station_id'].isin([41660])]
max_sum_pf.plot(x = 'date', y = 'rides', figsize=(20, 5))
```

```python
pf_ct=pf[pf['daytype'] == 'W'].groupby('station_id').sum()
pf_ct[pf_std['rides'] == min(pf_std['rides'])]
```

```python
min_sum_pf = pf[pf['station_id'].isin([41580])]
min_sum_pf.plot(x = 'date', y = 'rides', figsize=(20, 5))
```

```python
random_pf = pf[pf['station_id'].isin([40040])]
random_pf.loc[:, 'moving_avg']=random_pf['rides'].rolling(window = 60, center = False).mean()
random_pf.plot(x = 'date', y = 'moving_avg', figsize=(20, 5))
```
